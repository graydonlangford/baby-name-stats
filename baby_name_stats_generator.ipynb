```python
import pandas as pd
import glob
import re
```

```python
birth_data_folder = './your-birth-data-folder-here'
death_table_folder = './your-death-table-folder-here'
```

```python
# Import babies born in each year
birth_year_filenames = glob.glob(birth_data_folder + '/yob*.txt')

birth_year_list = []

for filename in birth_year_filenames:
    result = re.search(r'yob([0-9]{4})\.txt', filename)
    year = int(result.group(1))

    birth_frequencies = pd.read_csv(filename, sep=',', header=None)
    birth_frequencies.columns = ['Name','Sex','Born']
    
    # total_births_in_year = birth_frequencies['Born'].sum()
    birth_frequencies['Birth Year']=year
    # birth_frequencies['Born (Percent)'] = birth_frequencies['Born (Count)'] / total_births_in_year
    birth_year_list.append(birth_frequencies)

birth_frequencies = pd.concat(birth_year_list, axis=0, ignore_index=True)
# birth_frequencies.query("Year>=1900", inplace=True)

# Set types and indexes
birth_frequencies[['Name', 'Sex']] = birth_frequencies[['Name', 'Sex']].astype('category')
birth_frequencies['Born'] = birth_frequencies['Born'].astype('int32')
birth_frequencies['Birth Year'] = birth_frequencies['Birth Year'].astype('int16')
# birth_frequencies.set_index(['Sex', 'Year'], inplace=True)
birth_frequencies
```

```python
# import actuarial death tables per year
death_table_filenames = glob.glob(death_table_folder + '/*.csv')

death_table_list = []

for filename in death_table_filenames:
    result = re.search(r'DeathProbsE_(.)_.*\.csv', filename)
    sex = result.group(1)

    death_table_single = pd.read_csv(filename, skiprows=1)
    death_table_single.insert(loc=0, column='Sex', value=sex)

    death_table_list.append(death_table_single)

death_table = pd.concat(death_table_list, axis=0, ignore_index=True)

# Set types and indexes)
death_table['Sex'] = death_table['Sex'].astype('category')
death_table['Year'] = death_table['Year'].astype('int16')
death_table.rename(columns={'Year': 'Birth Year'}, inplace=True)
# death_table['Birth Year'] = death_table['Birth Year'].astype('int16')


subset_1900 = death_table.loc[death_table['Birth Year'] == 1900]
# subset_1900
# subset_1900['Survival Year'] = subset_1900['Survival Year'] - 20

for year in range(1880, 1900):
    subset_1900['Birth Year'] = year
    death_table = pd.concat([death_table, subset_1900], axis=0, ignore_index=True)

death_table
```

```python
sex_and_year = death_table[["Sex", "Birth Year"]]
```

```python
survival_table = death_table.filter(regex=r'[0-9]').map(lambda x: 1-x)
survival_table
```

```python
cumulative_survival_table = survival_table.cumprod(axis=1)
cumulative_survival_table = pd.concat([sex_and_year, cumulative_survival_table], axis=1) 
cumulative_survival_table
```

```python
offset_survival_rates = cumulative_survival_table.filter(regex=r'[0-9]').shift(periods=1, axis=1)
offset_survival_rates['0'] = 1
death_probabilities = death_table.filter(regex=r'[0-9]').multiply(offset_survival_rates)

death_probabilities = pd.concat([sex_and_year, death_probabilities], axis=1)
death_probabilities
```

```python
death_table_unpivot = death_probabilities.melt(id_vars=['Sex', 'Birth Year'], var_name="Age", value_name="Death Rate")
death_table_unpivot['Age'] = pd.to_numeric(death_table_unpivot['Age'])
death_table_unpivot['Target Year'] = death_table_unpivot['Birth Year'] + death_table_unpivot['Age']
death_table_unpivot = death_table_unpivot.loc[death_table_unpivot['Target Year'] <= 2023]
death_table_unpivot
```

```python
cumulative_survival_table_unpivot = cumulative_survival_table.melt(id_vars=['Sex', 'Birth Year'], var_name="Age", value_name="Cumulative Survival Rate")
cumulative_survival_table_unpivot['Age'] = pd.to_numeric(cumulative_survival_table_unpivot['Age'])
cumulative_survival_table_unpivot['Target Year'] = cumulative_survival_table_unpivot['Birth Year'] + cumulative_survival_table_unpivot['Age']
cumulative_survival_table_unpivot = cumulative_survival_table_unpivot.loc[cumulative_survival_table_unpivot['Target Year'] <= 2023]
cumulative_survival_table_unpivot
```

```python
death_survival_table = pd.concat([death_table_unpivot, cumulative_survival_table_unpivot["Cumulative Survival Rate"]], axis=1) 
death_survival_table
```

```python
death_survival_table_with_rates = birth_frequencies.merge(death_survival_table[['Sex', 'Birth Year', 'Age', 'Target Year', 'Death Rate', 'Cumulative Survival Rate']], how='left', left_on=['Sex', 'Birth Year'], right_on=['Sex', 'Birth Year'])
death_survival_table_with_rates
```

```python
death_survival_amounts = death_survival_table_with_rates
death_survival_amounts['Died'] = death_survival_amounts['Born'] * death_survival_amounts['Death Rate']
death_survival_amounts['Living'] = death_survival_amounts['Born'] * death_survival_amounts['Cumulative Survival Rate']
death_survival_amounts['Years Lived'] = death_survival_amounts['Living'] * death_survival_amounts['Age']
death_survival_amounts.drop(['Death Rate', 'Age', 'Cumulative Survival Rate'], axis=1, inplace=True)
death_survival_amounts = death_survival_amounts.round(0)
death_survival_amounts[['Died', 'Living']] = death_survival_amounts[['Died', 'Living']].astype('int32')
death_survival_amounts
```

```python
# newdf = death_survival_amounts.groupby(['Name', 'Sex', 'Target Year'], observed=True)[['Died', 'Living']].sum().reset_index()
died_and_living_per_year = death_survival_amounts.groupby(['Name', 'Sex', 'Target Year'], observed=True).agg(SumYears=('Years Lived', 'sum'), Died=('Died', 'sum'), Living=('Living', 'sum')).reset_index()
# died_and_living_per_year.rename(columns={'SumYears': 'Average Age'}, inplace=True)
died_and_living_per_year['Average Age'] = died_and_living_per_year['SumYears'] / died_and_living_per_year['Living']
died_and_living_per_year.drop('SumYears', axis=1, inplace=True)
died_and_living_per_year
```

```python
birth_frequencies
```

```python
all_baby_name_stats = died_and_living_per_year.merge(birth_frequencies, how='left', left_on=['Name', 'Sex', 'Target Year'], right_on=['Name', 'Sex', 'Birth Year'])[['Name', 'Sex', 'Target Year', 'Born', 'Died', 'Living', 'Average Age']]
all_baby_name_stats['Born'] = all_baby_name_stats['Born'].fillna(0).astype('int32')

all_baby_name_stats
```

```python
all_baby_name_stats.to_csv("./baby_name_stats.csv", header=['Name', 'Sex', 'Year', 'Born', 'Died', 'Living', 'Average Age'], index=False)
```
